# Caminante aleatorio
Este consiste en lanzar una moneda a cada paso y tomar una decisi√≥n hacia donde caminar
 Lo que hace es hechar un volado y caminar hacia arriba o hacia abajo

```python
from pylab import *
%matplotlib inline
```

```python
trayectoria = zeros(1000000)
posicion= 0
for n in range (len(trayectoria)):
    trayectoria[n] =posicion
    volado = rand()
    if volado < 0.5:
        posicion = posicion - 1
    else:
        posicion = posicion + 1
plot(range(len(trayectoria)),trayectoria)
show()
```
